Different age groups in the 'persons' table.

```python
sql_txt = r'''
SELECT 
    p1.age AS "AGE"
    , COUNT(1) AS "COUNT"
FROM persons p1
GROUP BY p1.age
ORDER BY COUNT(1) DESC
'''

HP.sql_to_df(db_path=db_path, sql_txt=sql_txt)
```

<br>

Age group with the maximum number of people.

```python
sql_txt = r'''
SELECT 
    MAIN.AGE
    , MAIN.AGE_COUNT
    
FROM ( 
            SELECT 
                AGE_COUNT.AGE
                , AGE_COUNT.AGE_COUNT
                , MAX(AGE_COUNT.AGE_COUNT) OVER() AS "MAX_AGE_COUNT"
            FROM ( 
                        SELECT 
                            p1.age AS "AGE"
                            , COUNT(1) AS "AGE_COUNT"
                        FROM persons p1
                        GROUP BY p1.age
                ) AGE_COUNT
    ) "MAIN" 

WHERE 
    "MAIN".AGE_COUNT = "MAIN".MAX_AGE_COUNT

'''

HP.sql_to_df(db_path=db_path, sql_txt=sql_txt)
```

<br>

Get all people that do not have a last name.

```python
sql_txt = r'''
SELECT 
    p1.*
FROM persons p1
WHERE 
    (
        p1.last_name IS NULL 
        OR p1.last_name = 'None'
        OR p1.last_name = ''
    )
'''

HP.sql_to_df(db_path=db_path, sql_txt=sql_txt)
```

<br>

Owners that have more than 1 pet.

```python
sql_txt = r'''
SELECT * FROM ( 
SELECT 
    PET_OWNER.id AS "OWNER_ID" 
    , PET_OWNER.first_name AS "OWNER_FIRST_NAME"
    , PET_OWNER.last_name AS "OWNER_LAST_NAME"
    , PET_OWNER.age AS "OWNER_AGE"
    , PET_OWNER.city AS "OWNER_CITY"
    , PET_OWNER.zip_code AS "OWNER_ZIP"
    , PETS.pet_name AS "PET_NAME"
    , PETS.pet_type AS "PET_TYPE"
    , PETS.treatment_done AS "TREATMENT_DONE"
    , COUNT(1) OVER(PARTITION BY PET_OWNER.id) AS "PET_COUNT"
FROM persons PET_OWNER
LEFT JOIN pets PETS ON PETS.owner_id = PET_OWNER.id
) "MAIN"
WHERE "MAIN".PET_COUNT > 1
'''

HP.sql_to_df(db_path=db_path, sql_txt=sql_txt)
```

<br>

Number of pets that have received treatment.

```python
sql_txt = r'''
SELECT 
    COUNT(1) AS "RECVD_TRTMNT_COUNT"
FROM pets PETS
WHERE PETS.treatment_done > 0
'''

HP.sql_to_df(db_path=db_path, sql_txt=sql_txt)
```

```python
sql_txt = r'''
SELECT 
    COUNT(1) AS "RECVD_TRTMNT_COUNT"
FROM pets PETS
WHERE 
    PETS.treatment_done > 0
    AND NOT ( 
        PETS.pet_type = 'None'
        OR PETS.pet_type = ''
        OR PETS.pet_type IS NULL 
    )
'''

HP.sql_to_df(db_path=db_path, sql_txt=sql_txt)
```

<br>

Number of pets where owner city = 'east port'.

```python
sql_txt = r'''
SELECT COUNT(1) AS "EAST_PORT_PET_COUNT" FROM ( 
SELECT 
    PET_OWNER.id AS "OWNER_ID" 
    , PET_OWNER.first_name AS "OWNER_FIRST_NAME"
    , PET_OWNER.last_name AS "OWNER_LAST_NAME"
    , PET_OWNER.age AS "OWNER_AGE"
    , PET_OWNER.city AS "OWNER_CITY"
    , PET_OWNER.zip_code AS "OWNER_ZIP"
    , PETS.pet_name AS "PET_NAME"
    , PETS.pet_type AS "PET_TYPE"
    , PETS.treatment_done AS "TREATMENT_DONE"
FROM persons PET_OWNER
LEFT JOIN pets PETS ON PETS.owner_id = PET_OWNER.id
) "MAIN"
WHERE "MAIN".OWNER_CITY = 'east port'
'''


HP.sql_to_df(db_path=db_path, sql_txt=sql_txt)
```

Number of pets where owner city = 'east port' and treatment was received.

```python
sql_txt = r'''
SELECT COUNT(1) AS "EAST_PORT_PET_COUNT" FROM ( 
SELECT 
    PET_OWNER.id AS "OWNER_ID" 
    , PET_OWNER.first_name AS "OWNER_FIRST_NAME"
    , PET_OWNER.last_name AS "OWNER_LAST_NAME"
    , PET_OWNER.age AS "OWNER_AGE"
    , PET_OWNER.city AS "OWNER_CITY"
    , PET_OWNER.zip_code AS "OWNER_ZIP"
    , PETS.pet_name AS "PET_NAME"
    , PETS.pet_type AS "PET_TYPE"
    , PETS.treatment_done AS "TREATMENT_DONE"
FROM persons PET_OWNER
LEFT JOIN pets PETS ON PETS.owner_id = PET_OWNER.id
) "MAIN"
WHERE 
    "MAIN".OWNER_CITY = 'east port'
    AND 
    "MAIN".TREATMENT_DONE > 0
'''


HP.sql_to_df(db_path=db_path, sql_txt=sql_txt)
```
