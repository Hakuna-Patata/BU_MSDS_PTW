```python
#nested 'while' loop example

#Get the mean of input numbers
n_numbers = int(input("How many numbers will be given? "))
sum_numbers = 0
count = 0

while count < n_numbers:
    current_number = input('Provide a number: ')
    while not current_number.isnumeric():
        current_number = input('That\'s not a number! Try again: ')
    sum_numbers += int(current_number)
    count += 1

print(sum_numbers / n_numbers)
```

```python
#'for' loop with break

#determines if the number entered is a prime number or not. will loop
#and divide user number by each number leading up to the user number and
#if there is a number that has no remainder after division, then loop will
#break and return false. If no divisions return a remainder of 0 then the 
#user number is a prime number and variable holding True will remain the same.
user_number = int(input('Please provide a number from 2 to 1000: '))
if user_number < 2 or user_number > 1000:
    print('Incorrect number!')
else:
    is_prime = True
    for digit in range(2, user_number):
        if user_number % digit == 0:
            is_prime = False
            break

if is_prime == False:
    print(user_number, 'is not prime!')
elif is_prime == True:
    print(user_number, 'is prime!')
```

```python
#'while' loop: using continue to skip loop iterations based on met criteria

#print only even numbers less than 50 that are NOT divisible by 6
current_value = 0
while current_value < 50:
    current_value += 2
    if current_value % 6 == 0:
        continue #will skip iteration if divisible by 6
    print(current_value)
```

```python
#using 'for' loop to get factorial of number

#A factorial is a number multiplied by all of the positive integers leading up to that number and
#the number itself.
#example: factorial of 5! = 5 * 4 * 3 * 2 * 1   =  120
iterations = 5
iteration_num = 1
for i in range(1, iterations+1):
    print(i,'x',iteration_num,'=',i*iteration_num)
    iteration_num = iteration_num * i
print('The factorial of',iterations,'is',iteration_num)
```

```python
#using nested 'for' loop to create patterns.
for val1 in range(1,9):
    string = ''
    for val2 in range(1,50):
        val2_1 = '~'
        val2_2 = '|'
        if val1 % 2 != 0:
            if val2 % 2 != 0:
                string = string + val2_1
            else:
                string = string + val2_2
        else:
            if val2 % 2 != 0:
                string = string + val2_2
            else:
                string = string + val2_1
    print(string)
```

```python

apartments = [
  ('12 Park Ave', 2, 78.5, 475000),
  ('15 Washington St', 3, 97.0, 685000),
  ('27 Lake St', 1, 46.7, 380000),
  ('3 Lake Ave', 2, 62.3, 420000)
]


def get_biggest_apartment_area_price(apartments):
    key = 0 
    apt_dict = {}
    for apt in apartments:
        apt_dict[key] = apt[2] 
        key += 1 
    area, price = apartments[max(apt_dict, key=apt_dict.get)][2:4]
    return (area, price)
  
get_biggest_apartment_area_price(apartments)

```

### <font color=1de30b>Sets</font>
---

Sets are similar to lists but all values within a set must be unique values. Elements can be added or removed from a set but the existing elements cannot be modified. Also, sets are unordered meaning you cannot access elements within them via an index. This also means that the order can sporadically change from execution to execution.

Turn list with duplicate values into set with unique values.

```python
num_list = [1,1,1,2,2,3,3,3,3,3,3,4]

num_set = set(num_list)

print(num_set)
```

<br>

<font color="yellow">
Add/Remove elements from a set.
</font>

```python
num_set = {1,2,3,4,5,6,7}

# Add element to a set
num_set.add(8)

# Remove element from a set. No exception thrown if doesn't exist.
num_set.discard(1)

# Remove element from set if exists else, throw exception
x = input("Enter number to remove: ")
try:
    num_set.remove(int(x))
except:
    print("Number not found in set!")
    
    
print(num_set)


# Remove all elements from a set
num_set.clear()
print(num_set)
```

<br>

<font color="yellow">
Get similarities between sets.
</font>

```python
num_set_1 = {1,2,3,4,5}

num_set_2 = {4,5,6,7,8}

# Using intersection method
print(num_set_1.intersection(num_set_2))

# Using syntax '&' for intersection method
print(num_set_1 & num_set_2)
```

<br>

<font color="yellow">
Combine unique values from sets into 1 set
</font>

```python
num_set_1 = {1,2,3,4,5}

num_set_2 = {4,5,6,7,8}

# Using union method
print(num_set_1.union(num_set_2))

# Using syntax '|' for union method
print(num_set_1 | num_set_2)
```

<br>

<font color="yellow">
Get unique differences between sets.
</font>

```python
num_set_1 = {1,2,3,4,5}

num_set_2 = {4,5,6,7,8}

# Using difference method
# Gets unique differences in first called set
print(num_set_1.difference(num_set_2))

# Using syntax  '-' for difference method
# Gets unique differences in first called set
print(num_set_2 - num_set_1)


# Using symmetric_difference method
# Gets unique differences in both sets
print(num_set_1.symmetric_difference(num_set_2))


# Using syntax '^' for symmetric_difference_method
print(num_set_1 ^ num_set_2)
```

<br>

<font color="yellow">
<ul>
    <li> Subset: A set in which all elements exist in another set with equal or more elements in it.
    <li> Superset: A set in which all elements exist in another set but has equal or more elements than the other set.
</ul>
</font>

```python
sub_set = {1,2,3,4}
sup_set = {1,2,3,4,5,6,7}
set1 = {1,2,3,4}
null_set = set({}) # without set(), would create an empty dictionary

print(sub_set.issuperset(sup_set))  # Less elements so False
print(sub_set.issubset(sup_set))  # Less elements so True
print(sub_set.issuperset(set1))  # If same number of elements then True
print(sub_set.issubset(set1))  # If same number of elements then True
print(null_set.issuperset(sub_set))  # If NULL set then any other non-NULL set will be superset
print(null_set.issubset(sub_set))  # If NULL set then is subset of any other set (including another null set)
print(null_set.issuperset(null_set))  # If same number of elements (none), then True
```

```python
import datetime

date_var = datetime.time(13, 55, 21, 5)  # date from parts (hour, minute, second, microsecond)

print(date_var)
print(date_var.hour)  # Get hour part of date object
print(date_var.minute)  # Get minute part of date object
print(date_var.second)  # Get second part of date object
print(date_var.microsecond)  # Get microsecond part of date object
```
